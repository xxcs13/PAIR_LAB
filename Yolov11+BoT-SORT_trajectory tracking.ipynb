# 5. Python Usage

YOLO11 was reimagined using Python-first principles for the most seamless Python YOLO experience yet. YOLO11 models can be loaded from a trained checkpoint or created from scratch. Then methods are used to train, val, predict, and export the model. See detailed Python usage examples in the [YOLO11 Python Docs](https://docs.ultralytics.com/usage/python/).

```
from ultralytics import YOLO

# Load a model
model = YOLO('yolo11n.yaml')  # build a new model from scratch
model = YOLO('yolo11n.pt')  # load a pretrained model (recommended for training)

# Use the model
results = model.train(data='coco8.yaml', epochs=3)  # train the model
results = model.val()  # evaluate model performance on the validation set
results = model('https://ultralytics.com/images/bus.jpg')  # predict on an image
results = model.export(format='onnx')  # export the model to ONNX format
```

```
import matplotlib.pyplot as plt

# Constants for the badminton court dimensions in meters (as per BWF regulations)
COURT_LENGTH = 13.4  # Total length
COURT_WIDTH = 6.1    # Total width
SINGLES_WIDTH = 5.18 # Singles court width
SERVICE_LINE_DISTANCE = 1.98 # Distance of the short service line from the net
BACK_BOUNDARY_DISTANCE = 0.76 # Distance of the back doubles service line from the back boundary line
CENTER_LINE = COURT_WIDTH / 2

# Convert meters to centimeters for better visualization scale
scale = 100  # 1 meter = 100 centimeters

# Calculate corner coordinates (in centimeters)
corners = [
    (0, 0),
    (COURT_WIDTH * scale, 0),
    (COURT_WIDTH * scale, COURT_LENGTH * scale),
    (0, COURT_LENGTH * scale)
]

# Draw the badminton court
def draw_modified_badminton_court():
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.set_aspect('equal', adjustable='box')

    # Line width for the court lines
    line_width = 1

    # Draw outer boundary
    ax.plot([0, 0, COURT_WIDTH * scale, COURT_WIDTH * scale, 0],
            [0, COURT_LENGTH * scale, COURT_LENGTH * scale, 0, 0], color='black', linewidth=line_width)

    # Draw singles side lines
    ax.plot([((COURT_WIDTH - SINGLES_WIDTH) / 2) * scale] * 2, [0, COURT_LENGTH * scale],
            color='black', linewidth=line_width)
    ax.plot([(COURT_WIDTH - ((COURT_WIDTH - SINGLES_WIDTH) / 2)) * scale] * 2, [0, COURT_LENGTH * scale],
            color='black', linewidth=line_width)

    # Draw net
    ax.plot([0, COURT_WIDTH * scale], [(COURT_LENGTH / 2) * scale] * 2, color='black', linewidth=line_width)

    # Draw short service lines
    ax.plot([0, COURT_WIDTH * scale], [(COURT_LENGTH / 2 - SERVICE_LINE_DISTANCE) * scale] * 2,
            color='black', linewidth=line_width)
    ax.plot([0, COURT_WIDTH * scale], [(COURT_LENGTH / 2 + SERVICE_LINE_DISTANCE) * scale] * 2,
            color='black', linewidth=line_width)

    # Draw center line (excluding the part between short service line and net)
    ax.plot([CENTER_LINE * scale] * 2, [0, (COURT_LENGTH / 2 - SERVICE_LINE_DISTANCE) * scale],
            color='black', linewidth=line_width)
    ax.plot([CENTER_LINE * scale] * 2, [(COURT_LENGTH / 2 + SERVICE_LINE_DISTANCE) * scale, COURT_LENGTH * scale],
            color='black', linewidth=line_width)

    # Draw doubles service lines (back boundary lines)
    ax.plot([0, COURT_WIDTH * scale], [(COURT_LENGTH - BACK_BOUNDARY_DISTANCE) * scale] * 2,
            color='black', linewidth=line_width)
    ax.plot([0, COURT_WIDTH * scale], [BACK_BOUNDARY_DISTANCE * scale] * 2,
            color='black', linewidth=line_width)

    # Adjust axis to remove labels and ticks
    ax.axis('off')

    # Save the court image
    plt.savefig('/content/drive/MyDrive/badminton_court.png', bbox_inches='tight', pad_inches=0, dpi=300)
    plt.close()

    # Save corner coordinates to a text file
    with open('/content/drive/MyDrive/court_corners.txt', 'w') as file:
        for i, (x, y) in enumerate(corners):
            file.write(f"Corner {i + 1}: ({x:.1f} cm, {y:.1f} cm)\n")

# Execute the function
draw_modified_badminton_court()
```

## ËªåË∑°ÊäïÂΩ±Ëá≥Âπ≥Èù¢ÁêÉÂ†¥

```
from collections import defaultdict, deque
import cv2
import numpy as np
from PIL import Image, ImageDraw
import time
from ultralytics import YOLO

# Load the YOLO model
model = YOLO("yolo11n.pt")

# Open the video file
video_path = "/content/drive/MyDrive/tmp0.mp4"
cap = cv2.VideoCapture(video_path)

# Load the flat court image
court_image_path = "/content/drive/MyDrive/badminton_court.png"
flat_court = cv2.imread(court_image_path)
court_height, court_width = flat_court.shape[:2]

# Define the court corners in the video and the flat court
video_court_corners = np.array([(340, 60), (134, 443), (700, 455), (513, 60)], dtype=np.float32)
flat_court_corners = np.array([(0, 0), (0, 1340), (610, 1340), (610, 0)], dtype=np.float32)

# Compute the homography matrix
homography_matrix, _ = cv2.findHomography(video_court_corners, flat_court_corners)

# Store track history for each ID (use deque for limited length)
track_history = defaultdict(lambda: deque(maxlen=8))

# Prepare the output video writer
output_path = "/content/drive/MyDrive/court_trajectory.mp4"
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
fps = int(cap.get(cv2.CAP_PROP_FPS))
out = cv2.VideoWriter(output_path, fourcc, fps, (court_width, court_height))

start_time = time.time()

# Loop through the video frames
while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    # Run YOLO tracking on the frame
    results = model.track(frame, persist=True)

    # Get the boxes, class IDs, and track IDs
    boxes = results[0].boxes.xywh.cpu()
    class_ids = results[0].boxes.cls.int().cpu().tolist()
    track_ids = results[0].boxes.id.int().cpu().tolist()

    # Create a copy of the flat court to draw trajectories
    current_court = flat_court.copy()

    # Process each detection
    for box, class_id, track_id in zip(boxes, class_ids, track_ids):
        if class_id == 0:  # Only process people
            x, y, w, h = box
            center_point = np.array([[int(x), int(y)]], dtype=np.float32)

            # Transform the point to flat court coordinates
            transformed_point = cv2.perspectiveTransform(center_point[None, :, :], homography_matrix)[0][0]
            flat_point = (int(transformed_point[0]), int(transformed_point[1]))

            # Update track history
            track_history[track_id].append(flat_point)

            # Draw trajectory on the current court
            for point in track_history[track_id]:
                cv2.circle(current_court, point, 15, (0, 0, 255), -1)

    # Write the updated court frame to the output video
    out.write(current_court)

end_time = time.time()
total_time = end_time - start_time

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()

# Print the inference time
print(f"Total inference time: {total_time:.2f} seconds")
```

## Âú®ÂéüÂßãÂΩ±Áâá‰∏≠Áï´Âá∫ËªåË∑°

```
from collections import defaultdict, deque
import cv2
import numpy as np
from PIL import Image, ImageDraw
import time
from ultralytics import YOLO

# Load the YOLO11 model
model = YOLO("yolo11n.pt")

# Open the video file
video_path = "/content/drive/MyDrive/tmp0.mp4"
cap = cv2.VideoCapture(video_path)

# Get video properties
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
fourcc = cv2.VideoWriter_fourcc(*"mp4v")  # Codec for MP4 format

# Define the output video writer
output_path = "/content/drive/MyDrive/tmp3_traj.mp4"
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

# Function to check if a point is inside a polygon
def is_point_in_polygon(point, polygon):
    return cv2.pointPolygonTest(np.array(polygon, dtype=np.int32), point, False) >= 0

# Define the court corners based on the provided coordinates
court_corners = [
    (340, 60),   # Left-top(V)(340, 80)
    (513, 60),   # Right-top(513, 60)
    (700, 456),  # Right-bottom(700, 456)
    (134, 444)   # Left-bottom(134, 444)
]

# Store track history for each ID (use deque for limited length)
track_history = defaultdict(lambda: deque(maxlen=8))

start_time = time.time()

# Loop through the video frames
current_frame = 0
while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    # Run YOLO11 tracking on the frame, persisting tracks between frames
    results = model.track(frame, persist=True)

    # Get the boxes, class IDs, and track IDs
    boxes = results[0].boxes.xywh.cpu()
    class_ids = results[0].boxes.cls.int().cpu().tolist()
    track_ids = results[0].boxes.id.int().cpu().tolist()

    # Draw the court area for visualization
    # cv2.polylines(frame, [np.array(court_corners, dtype=np.int32)], isClosed=True, color=(255, 255, 0), thickness=2)

    # Convert frame to PIL for drawing
    pil_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    pil_frame = Image.fromarray(pil_frame)

    # Process each detection
    for box, class_id, track_id in zip(boxes, class_ids, track_ids):
        if class_id == 0:  # Only process people
            x, y, w, h = box
            center_point = (int(x), int(y))

            # Check if the person is inside the court
            if is_point_in_polygon(center_point, court_corners):
                # Update track history
                track_history[track_id].append(center_point)

                # Draw trajectory
                for point in track_history[track_id]:
                    bbox = (point[0] - 2, point[1] - 2, point[0] + 2, point[1] + 2)
                    draw = ImageDraw.Draw(pil_frame)
                    draw.ellipse(bbox, outline='red')
                    del draw

    # Convert PIL frame back to OpenCV
    frame = cv2.cvtColor(np.array(pil_frame), cv2.COLOR_RGB2BGR)

    # Write the annotated frame to the output video
    out.write(frame)

    current_frame += 1

end_time = time.time()
total_time = end_time - start_time

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()

# Print the inference time and speed
print(f"Total inference time: {total_time:.2f} seconds")
```
